```python
# Import the necessary libraries
import cv2
import matplotlib.pyplot as plt

# Load the image and convert to rbg
image1 = cv2.imread('img/Adaora-Onaga.jpg')
image = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)


# Plot the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image)

# Scale the image by a factor of 2 along both axes
scaled_image = cv2.resize(image, None, fx=2, fy=2)

# Save the image
cv2.imwrite('img/Adaora-Onaga-image-Scaled.jpg', scaled_image)

# Plot the scaled image
plt.subplot(1, 2, 2)
plt.title("Scaled")
plt.imshow(scaled_image)
plt.show()
```
