```python
# Import the necessary libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load the image
image = cv2.imread('img/cookies.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Plot the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image_rgb)

# Inverse by subtracting from 255
inverse_image = 255 - image_rgb

# Save the image
cv2.imwrite('img/inverse_image.jpg', inverse_image)

# Plot the inverse image
plt.subplot(1, 2, 2)
plt.title("Inverse color")
plt.imshow(inverse_image)
plt.show()
```
