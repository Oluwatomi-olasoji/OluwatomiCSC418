```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('img/cookies.jpg')
image_rbg = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(image_rbg)

#removing the noise and saving the image
filteredimg = cv2.medianBlur(image_rbg, 15)
cv2.imwrite('img/cookies-Meidan-blur.jpg',filteredimg)

plt.subplot(1,2,2)
plt.title("Blurred cookeis")
plt.imshow(filteredimg)
plt.show()
```
