```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('img/grayImage.jpg')
```

```python
plt.subplot(1,2,1)
plt.title("original")
plt.imshow(image)
```

```python
brightness = 8
contrast = 2
image2 = cv2.addWeighted(image, contrast, np.zeros(image.shape,image.dtype),0, brightness)
```

```python
cv2.imwrite('img/contrast_image.jpg',image2)
```

```python
plt.subplot(1,2,2)
plt.title("Brightness & contrast")
plt.imshow(image2)
plt.show()
```
