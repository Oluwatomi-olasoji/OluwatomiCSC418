```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
image = cv2.imread('img/Sola-Oni.jpg')
```

```python
plt.subplot(1,2,1)
plt.title("Original")
# Convert BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
```

```python
#sharpening kernel
kernel = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])

#sharpen the image
sharp = cv2.filter2D(image_rgb,-1,kernel)

#save
cv2.imwrite('img/sharp_DrSola.jpg',sharp)

#plot the sharpened image
plt.subplot(1,2,2)
plt.title("sharpening")
plt.imshow(sharp)
plt.show()
```
