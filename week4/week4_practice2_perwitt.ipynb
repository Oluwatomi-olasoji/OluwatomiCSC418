```python
# perwitt edge detection

import cv2
import numpy as np
from google.colab.patches import cv2_imshow # Import the cv2_imshow function
from matplotlib import pyplot as plt # Import pyplot for displaying with titles


# Load the image
image = cv2.imread('/Sola-Oni.jpg', cv2.IMREAD_GRAYSCALE)

# resize images
image= cv2.resize(image, (500,400))

# Define Prewitt kernels
kernel_x = np.array([[-1,  0,  1],
                     [ 0,  0,  0],
                     [ 1,  0,  1]])

kernel_y = np.array([[-1,  0,  1],
                     [-1,  0,  1],
                     [-1,  0,  1]])

# Apply Prewitt edge detection
prewitt_x = cv2.filter2D(image, -1, kernel_x)
prewitt_y = cv2.filter2D(image, -1, kernel_y)

# Compute magnitude of gradients manually
prewitt_combined = np.sqrt(np.square(prewitt_x) + np.square(prewitt_y))

# Display the original image and the Prewitt edges
# cv2_imshow('Original', image) # This line caused the error
# cv2_imshow('Prewitt Edges', np.uint8(prewitt_combined)) # This line caused the error

# Display using matplotlib for titles
plt.figure(figsize=(10, 5)) # Create a figure to hold the plots
plt.subplot(1, 2, 1) # Create a subplot for the original image
plt.imshow(image, cmap='gray') # Display the image using imshow
plt.title('Original') # Set the title of the subplot
plt.axis('off') # Turn off axis labels and ticks

plt.subplot(1, 2, 2) # Create a subplot for the Prewitt edges
plt.imshow(np.uint8(prewitt_combined), cmap='gray') # Display the edges using imshow
plt.title('Prewitt Edges') # Set the title of the subplot
plt.axis('off') # Turn off axis labels and ticks

plt.show() # Show the figure with both subplots


# Wait for a key press and close all windows
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```
