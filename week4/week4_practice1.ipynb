```python
# Sobel edge detection

import cv2
import numpy as np
from google.colab.patches import cv2_imshow # Import the cv2_imshow function


# Load the image
image = cv2.imread('/Sola-Oni.jpg', cv2.IMREAD_GRAYSCALE)

# resize images
image= cv2.resize(image, (500,400))

# Apply Sobel edge detection
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

# Combine the gradient images
sobel_combined = cv2.magnitude(sobel_x, sobel_y)

# Display the original image and the Sobel edges
cv2_imshow(image)  # Use cv2_imshow instead of cv2.imshow
cv2_imshow(np.uint8(sobel_combined)) # Use cv2_imshow instead of cv2.imshow


# Wait for a key press and close all windows
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```
