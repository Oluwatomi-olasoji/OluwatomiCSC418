```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
img = cv.imread ( 'img/download.jpg' , 0 )

#Plot the original image
plt.subplot ( 1 , 2 , 1 )
plt.title ( " Original " )
plt.imshow ( img )

cropped_img = img[50:150, 50:100]

#Plot the original image
plt.subplot ( 1 , 2 , 2 )
plt.title ( " Cropped Image " )
plt.imshow ( cropped_img )

cv.imwrite ( 'cropped_out_downlaod.jpg' , cropped_img )
cv.waitKey ( 0 )
cv.destroyAllWindows ( )
```
